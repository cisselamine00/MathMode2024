## Exercise 1
1. Создайте равномерное разбиение интервала от -1.3 до 2.5 на 64 отрезка.  
2. Сгенерировать numpy массив длины $3n$, заполненный циклически числами 1, 2, 3, 1, 2, 3, 1....  
3. Создайте массив первых 10 нечетных целых чисел.  
4. Создайте массив нулей размера 10 x 10, а затем создайте в нём "рамку" из единиц по краям.  
5. Создайте массив 8 x 8 с шахматной доской из нулей и единиц.  
6. Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.


```python
import numpy as np

interval_values = np.linspace(-1.3, 2.5, 64)
n_value = 10
cyclic_numbers = np.tile(np.arange(1, 4), n_value)[:3 * n_value]
first_odd_numbers = np.arange(1, 20, 2)[:10]
zero_matrix = np.zeros((10, 10))
zero_matrix[0, :], zero_matrix[-1, :], zero_matrix[:, 0], zero_matrix[:, -1] = 1, 1, 1, 1
chess_board = np.zeros((8, 8), dtype=int)
chess_board[1::2, ::2] = 1
chess_board[::2, 1::2] = 1

def create_matrix(size):
    return np.fromfunction(lambda i, j: i + j, (size, size), dtype=int)

result_matrix = create_matrix(n_value)

print("Интервал значений:", interval_values)
print("Циклический массив (длина 3n):", cyclic_numbers)
print("Первые 10 нечетных чисел:", first_odd_numbers)
print("с рамкой из единиц:\n", zero_matrix)
print("Шахматы:\n", chess_board)
print(f"Матрица {n_value} x {n_value} с (i, j)-элементом i + j:\n", result_matrix)
```

## Exercise 2  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!
2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)
3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>! Выберите несколько интересных спец функций, которые вы знаете (или, может быть, не знаете), подставьте несколько чисел и начертите их.  
4. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.


```python
import matplotlib.pyplot as plt
from scipy.special import jn

x_1 = np.linspace(-1, 1, 250)
y_1 = x_1**2 * np.sin(1 / x_1**2) + x_1

plt.figure(figsize=(8, 5))
plt.plot(x_1, y_1)
plt.title('График')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

x_2 = np.linspace(5, 25, 100)
y_diff = np.abs(1 / (1 + x_2**2) - 1 / x_2**2)

plt.figure(figsize=(8, 5))
plt.semilogy(x_2, y_diff, label='разница')
plt.title('разница функций')
plt.xlabel('x')
plt.ylabel('разница')
plt.legend()
plt.grid(True)
plt.show()

x_3 = np.linspace(0, 5, 100)
y_bessel = jn(2, x_3)

plt.figure(figsize=(8, 5))
plt.plot(x_3, y_bessel, label=r'$J_2(x)$')
plt.title('График')
plt.xlabel('x')
plt.legend()
plt.grid(True)
plt.show()

x_mesh, y_mesh = np.meshgrid(np.linspace(-np.pi, np.pi, 100), np.linspace(-np.pi, np.pi, 100))
z_mesh = np.sin(x_mesh) * np.sin(y_mesh)

plt.figure(figsize=(8, 6))
plt.pcolormesh(x_mesh, y_mesh, z_mesh, cmap='viridis')
plt.colorbar(label='sin(x) * sin(y)')
plt.title('sin(x) * sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
