```python
from PIL import Image
import numpy as np

# Чтение изображения
img = Image.open('/content/téléchargement.jpeg')
# Преобразование в массив numpy
X = np.array(img, dtype=np.float32)

# Проверка, является ли изображение оттенками серого или цветным
if len(X.shape) == 2:  # Если изображение в оттенках серого
    U, S, V = np.linalg.svd(X, full_matrices=False)
    for r in [1, 2, 10, 30, 100]:
        # Восстановление изображения с ограниченным количеством сингулярных значений
        сжатое_изображение = np.dot(U[:, :r], np.dot(np.diag(S[:r]), V[:r, :]))
        # Обратное преобразование массива numpy в изображение PIL и сохранение изображения на диск
        Image.fromarray(np.asarray(сжатое_изображение, dtype=np.uint8)).save(f'сжатое_изображение_{r}.png')
        # Расчет размера сжатого изображения
        размер_сжатого = сжатое_изображение.nbytes / 1024
        print(f"Размер сжатого изображения с {r} сингулярными значениями: {размер_сжатого} Кб")
else:  # Если изображение цветное (RGB)
    # Применение SVD к каждому каналу отдельно для цветного изображения RGB
    U_r, S_r, V_r = np.linalg.svd(X[:,:,0], full_matrices=False)
    U_g, S_g, V_g = np.linalg.svd(X[:,:,1], full_matrices=False)
    U_b, S_b, V_b = np.linalg.svd(X[:,:,2], full_matrices=False)

    значения_сингулярных = [1, 2, 10, 30, 100]

    for r in значения_сингулярных:
        # Восстановление каждого канала с ограниченным количеством сингулярных значений
        сжатое_изображение_r = np.dot(U_r[:, :r], np.dot(np.diag(S_r[:r]), V_r[:r,:]))
        сжатое_изображение_g = np.dot(U_g[:, :r], np.dot(np.diag(S_g[:r]), V_g[:r,:]))
        сжатое_изображение_b = np.dot(U_b[:, :r], np.dot(np.diag(S_b[:r]), V_b[:r,:]))

        # Восстановление цветного изображения из сжатых каналов
        сжатое_изображение = np.zeros_like(X)
        сжатое_изображение[:,:,0] = сжатое_изображение_r
        сжатое_изображение[:,:,1] = сжатое_изображение_g
        сжатое_изображение[:,:,2] = сжатое_изображение_b

        # Обратное преобразование массива numpy в изображение PIL и сохранение изображения на диск
        Image.fromarray(np.asarray(сжатое_изображение, dtype=np.uint8)).save(f'сжатое_изображение_{r}.png')

        # Расчет размера сжатого изображения
        размер_сжатого = сжатое_изображение.nbytes / 1024
        print(f"Размер сжатого изображения с {r} сингулярными значениями: {размер_сжатого} Кб")
```
